```python
def main():
    with GraphDatabase.driver(URI, auth=AUTH) as driver:
        with driver.session(database="neo4j") as session:
            for state in zhvi['StateName'].unique():       
                session.execute_write(county_tx, state)
                print(f"State {state} added")


def county_tx(tx, name):
    # Create new County node with given name, if not exists already
    result = tx.run("""
        MERGE (s:State {name: $name})
        RETURN s.name AS name
        """, name=name
    )


main()
```

```python
def main():
    with GraphDatabase.driver(URI, auth=AUTH) as driver:
        with driver.session(database="neo4j") as session:
            for state in zhvi['StateName'].unique():       
                # Call the function to add counties for each state
                add_counties_for_state(session, state)
                print(f"Counties added for State {state}")


def add_counties_for_state(session, state_name):
    # Fetch county names for the given state
    state_filtered_data = zhvi[zhvi['State'] == state_name]
    county_names = state_filtered_data['RegionName'].unique()

    # Create "has county" relationships for the state
    for county_name in county_names:
        session.execute_write(add_county_relationship, state_name, county_name)


def add_county_relationship(tx, state_name, county_name):
    # Create a relationship "has county" between state and county
    tx.run("""
        MATCH (s:State {name: $state_name})
        MERGE (c:County {name: $county_name})
        MERGE (s)-[:HAS_COUNTY]->(c)
        """, state_name=state_name, county_name=county_name
    )


main()
```
